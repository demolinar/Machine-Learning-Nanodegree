```python
from numpy.core.fromnumeric import size
# import statement
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

# Load the data
bmi_life_data = pd.read_csv('bmi_and_life_expectancy.csv')

# Linear regression model
# Fit the model and Assign it to bmi_life_model
bmi_life_model = LinearRegression()
bmi_life_model.fit(bmi_life_data[['BMI']], bmi_life_data[['Life expectancy']])

# Predict life expectancy for 21.07931
print('Predict life expectancy for 21.07931 :',bmi_life_model.predict([[21.07931]]))

if __name__ == "__main__":
    # perform gradient descent
    X = bmi_life_data['BMI']
    y = bmi_life_data['Life expectancy']
    # plot the results
    plt.figure()
    plt.scatter(X, y)
    X = bmi_life_data['BMI'].values.reshape(-1, 1)
    
    plt.plot(X, bmi_life_model.predict(X),color='k')
    # plot the results of predict 21.07931 in color green
    plt.scatter(21.07931, bmi_life_model.predict([[21.07931]]),color='g',zorder = 3)
    plt.show()
```
